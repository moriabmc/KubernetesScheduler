# Imports and Methods

```python
import csv
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error as mape
import prophet

def read_co2_data(filePath):
    co2_emission_column = 3
    unix_timestamp_column = 1

    df = pd.DataFrame
    a = 0
    time = []
    emission = []
    timezone_unix_factor = 3600 # factor is necessary, as the unix timestamp is not standard compliant as timezone is set to UTC + 1
    with open(filePath, 'r') as csvfile:
        lines = csv.reader(csvfile, delimiter=',')
        next(lines)  # skip the label field
        for row in lines:
            a = a + 1
            day = datetime.datetime.fromtimestamp(
                int(row[unix_timestamp_column]) - timezone_unix_factor)  # correction as timestamp is UTC not UTC + 1
            time.append(day)
            try:
                emission.append(float(row[co2_emission_column]))
            except:
                emission.append(None)
    df = pd.DataFrame(list(zip(time, emission)), columns=['ds', 'y'])
    df['ds'] = pd.to_datetime(df['ds'])
    df = df.drop_duplicates(subset=['ds'])
    df.index = df['ds']
    df = df.drop(columns=['ds'])
    return df

def get_historical(df: pd.DataFrame, end_date: datetime.datetime):
    start_date = df.index[0]
    end_date -= datetime.timedelta(hours=1)
    daterange = pd.date_range(start_date, end_date, freq='H')
    result = df.loc[(df.index < end_date)]
    result = result.reindex(daterange, fill_value=np.nan)
    return result

def get_testdata(df: pd.DataFrame, start_date: datetime.datetime, end_date: datetime.datetime):
    daterange = pd.date_range(start_date, end_date, freq='H')
    result = df.loc[(df.index >= start_date ) & (df.index <= end_date)]
    result = result.reindex(daterange, fill_value=np.nan)
    return result
```

# Constants and Parameters

```python
FILENAME = 'Germany_CO2_Signal_2018.csv' # file containing the co2 data

# specify the timeframe, must be in iso format
FROM = '2018-11-24 00:00:00'
TO = '2018-11-24 23:00:00'

# Parameters for the prediction model
STEPS = 24 # number of hours predicted by the forecaster
```

# Read data

```python
df = read_co2_data(FILENAME)
df.head()
```

# Set a Timeframe

```python
# convert strings to datetime
frm = datetime.datetime.fromisoformat(FROM)
to = datetime.datetime.fromisoformat(TO)

df_hist = get_historical(df, frm)
df_hist.tail()
```

# Get Test Data

```python
df_test = get_testdata(df, frm, to)
df_test.head()
```

# Predict CO2

```python
df_hist['ds'] = df_hist.index
#df_hist = df_hist.interpolate(method='linear')

m = prophet.Prophet()
m.fit(df_hist)
future = m.make_future_dataframe(periods=STEPS, freq='H')
forecast = m.predict(future)
fig1 = m.plot(forecast)
```

Trends

```python
fig2 = m.plot_components(forecast)
```

```python
forecast.index = forecast['ds']
fig, ax = plt.subplots(figsize=(9, 4))
df_hist[-STEPS*2:]['y'].plot(ax=ax, label='train')
df_test['y'].plot(ax=ax, label='test')
forecast['yhat'][-STEPS:].plot(ax=ax, label='predictions')
ax.legend()
```

MAPE

```python
error = mape(df_test['y'], forecast['yhat'][-STEPS:])
print(error)
```

LME

```python
observed_min = df_test[['y']].idxmin()
predicted_min = forecast['yhat'][-STEPS:].idxmin()
print("lme", predicted_min - observed_min)
```

# Save prediction

```python
np.savetxt("co2_prediction.csv", forecast['yhat'][-STEPS:].to_numpy(), delimiter=",")
```
